```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import matplotlib.pyplot as plt

#pie chart showing most common star type in the data

def pie_star_commonality():
    
    data = pd.read_csv(os.path.join(dirname,filename))
    
    brown_dwarf = 0
    red_dwarf = 0
    white_dwarf = 0
    main_sequence = 0
    supergiant = 0
    hypergiant = 0
    d = {}
    for i in data['Star type']:
        if i == 0:
            brown_dwarf += 1
        elif i == 1:
            red_dwarf += 1
        elif i == 2:
            white_dwarf += 1
        elif i == 3:
            main_sequence += 1
        elif i == 4:
            supergiant += 1
        elif i == 5:
            hypergiant += 1           
    
    
    #Plotting a pie chart
    l1=['Brown Dwarf','Red_Dwarf', 'White_Dwarf', 'Main_Sequence', 'Supergiant', 'Hypergiant']
    number_of_stars = [brown_dwarf, red_dwarf, white_dwarf, main_sequence, supergiant, hypergiant]
    
    fig1, ax1 = plt.subplots()
    ax1.pie(number_of_stars, labels=l1,)

    plt.show()    
    
pie_star_commonality()


#Analysing the relation between temperature and spectral class

def bar_graph():

    data = pd.read_csv(os.path.join(dirname,filename))   
    
    
    grouper = data.groupby(['Spectral Class'])
    
    #Printing table with max temperature for each Spectral class
    print("DISPLAYING SPECTRAL CLASS AND MAX TEMPERATURE OF EACH CLASS")
    print(grouper['Temperature (K)'].max().reset_index())
    print('\n')
    
    #Printing table with min temperature for each Spectral class
    print("DISPLAYING SPECTRAL CLASS AND MIN TEMPERATURE OF EACH CLASS")
    print(grouper['Temperature (K)'].min().reset_index())
    print('\n')
    
    
    x = grouper['Temperature (K)'].max().reset_index()
    l1 = x.values.tolist()
    
    x_axis = []
    y_axis = []
    
    for i in l1:
        x_axis.append(i[0])
        y_axis.append(i[1])    
    
    
    y = grouper['Temperature (K)'].min().reset_index()
    l2 = y.values.tolist()
    
    x1_axis = []
    y1_axis = []
    
    for i in l2:
        x1_axis.append(i[0])
        y1_axis.append(i[1])
        
    #To obtain mid range temperature value for each spectral class
    mid_temp=[]
    
    for i in range(7):
        mid_temp.append((y_axis[i]+y1_axis[i])/2)

    
    #plotting bar graph for each spectral class and its max temperature
    
    plt.bar(x_axis,mid_temp)
    plt.xlabel("Spectral Class of Star")
    plt.ylabel("Temperature (K)")
    plt.title("Relation between spectral classes and temperature")
    plt.show()
    

bar_graph() 

 
#Scatterplot to show the luminosities of the various star types

def sctrplot():
    data = pd.read_csv(os.path.join(dirname,filename))

    x = np.array(data['Star type'])
    y = np.array(data['Luminosity(L/Lo)'])

    plt.scatter(x,y, color = 'red')
    plt.xlabel("Star type")
    plt.ylabel("Luminosity (L/Lo)")
    plt.title("The relation between star type and luminosity")
    plt.show()

                 
sctrplot()


    
    
            
   
    
```
